### Q.1) How Many Matches won by the Teams?

```python
plt.figure(figsize=(14,8))
ax=sns.countplot(x=df['Winner'])
plt.xticks(rotation=90)

# Add numbers on top of bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', 
                xytext=(0, 10), 
                textcoords='offset points')

plt.show()     #To Plot the Graph
```

### Q.2) What is the Percentage of the different ways of winning a match?

```python
plt.title('Result',fontsize=15)
df['Result'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()      #To Plot the Graph
```

### Q.3) What is the Distribution of Match Type, Toss Winner and Toss Decision?

```python
Matches = ['Match_Type', 'Toss_Winner', 'Toss_Decision']

# Define some different color palettes
Colour=['Green','BLue','Red']

for i, col in enumerate(Matches):
    plt.figure(figsize=(10, 4))
    ax=sns.countplot(x=df[col],order=df[col].value_counts().index,color=Colour[i % len(Colour)])
    plt.title(f"Distribution of {col}")
    plt.xticks(rotation=45)

# Add counts on top of bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}',
                 (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='bottom',
                 fontsize=9, color='black',
                 xytext=(0, 5), textcoords='offset points')
    
plt.show()    #To PLot the Graph
```

### Q.4) What is the Distribution of Result Margin, Target Runs and Target Overs?

## Bivariate and Multivariate Analysis

### Q.5) What is the relation between Target Runs vs Result Margin?

```python
# Scatter plot: Result Margin vs Target Runs
plt.figure(figsize=(12, 6))
plt.scatter(df['Result_Margin'],df['Target_Runs'],color='green',alpha=0.6)
plt.title('Target Runs vs Result Margin',fontsize=15,color='red')
plt.xlabel('Result Margin', fontsize=15)
plt.ylabel('Target Runs', fontsize=15)
plt.grid(True)
plt.show()    #To Plot the Graph
```

### Q.6) What is the Total Matches Played by City?

```python
# Create pivot-style data:Total Matches Played Per City
City_Matches=df['City'].value_counts()

# Plot
ax=City_Matches.plot(kind='bar', figsize=(12, 6), color='skyblue')
plt.title("Total Matches Played by City", fontsize=16)
plt.xlabel("City", fontsize=14)
plt.ylabel("Number of Matches", fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numbers on top of bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=10, 
                xytext=(0, 5), 
                textcoords='offset points')

plt.tight_layout()
plt.show()     #To Plot the Graph
```

### Q.7) What are the number of top 5 teams with most no. of final appearances?

```python
Finals_df=df[df['Match_Type']=='Final']

#Count appearances in Finals for both Team 1 and Team 2
Final_Teams=pd.concat([Finals_df['Team_1'],Finals_df['Team2']])
Team_Final_Counts=Final_Teams.value_counts()     # Get counts
  
# Get top 5 teams
Top_5_Teams=Team_Final_Counts.head(5)

# Plot the Graph
plt.figure(figsize=(10, 6))
ax=Top_5_Teams.plot(kind='bar',color='orange')
plt.title('Top 5 Teams with Most Final Appearances', fontsize=16)
plt.xlabel('Team', fontsize=14)
plt.ylabel('Number of Final Appearances', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add numbers on top of bars
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='bottom', 
                fontsize=10, 
                xytext=(0, 5), 
                textcoords='offset points')

plt.tight_layout()
plt.show()    #To Plot the Graph
```

### Q.8) What is the Correlation between Target Runs vs Result Margin in IPL Finals?

```python
Finals_df=df[df['Match_Type']=='Final']  # Filter only Final matches

# Scatter plot: Target Runs vs Result Margin
plt.figure(figsize=(10, 6))
plt.scatter(Finals_df['Target_Runs'],Finals_df['Result_Margin'],color='purple',alpha=0.7,edgecolors='k')
plt.title('Target Runs vs Result Margin in IPL Finals',fontsize=16)
plt.xlabel('Target Runs',fontsize=14)
plt.ylabel('Result Margin',fontsize=14)
plt.grid(True,linestyle='--',alpha=0.7)

z = np.polyfit(Finals_df['Target_Runs'],Finals_df['Result_Margin'], 1)   # Add trendline
p = np.poly1d(z)
plt.plot(Finals_df['Target_Runs'],p(Finals_df['Target_Runs']), "r--")  # Regression line
plt.tight_layout()
plt.show()     #To Plot the Graph
```

### Q.9) What are the no. of matches played between IPL Teams?

```python
# Create a pivot table: Number of matches between each pair of teams
Matches_between_Teams=pd.crosstab(df['Team_1'],df['Team2'])

# Plot the heatmap
plt.figure(figsize=(16, 10))
sns.heatmap(Matches_between_Teams,annot=True,fmt='d',cmap='YlGnBu',linewidths=0.5)

plt.title('Number of Matches Played Between Teams (IPL)', fontsize=18)
plt.xlabel('Team 2', fontsize=14)
plt.ylabel('Team 1', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()     #To Plot the Graph
```

### Q.10) What is the relation between Target Runs, Result Margin, Season, Super Over and Method?

```python
# Filter only Final matches
Finals_df=df[df['Match_Type']=='Final']

# Select more columns
Finals_Data=Finals_df[['Target_Runs','Result_Margin','Season','Super_Over','Method']]

# For method, fill NaN with 'Normal' (otherwise Seaborn will throw warnings)
Finals_Data['Method'].fillna('Normal', inplace=True)

# Plot the pairplot
sns.pairplot(Finals_Data,hue='Super_Over',diag_kind='kde',palette='husl',plot_kws={'alpha':0.7,'s':70,'edgecolor':'k'})

plt.suptitle('Expanded Pair Plot of IPL Finals Data', y=1.02, fontsize=16)
plt.show()      #To Plot the Graph
```

### Q.11) What is the relation between Toss Winner v/s Match Winner (IPl FInal)?

```python
from sklearn.preprocessing import LabelEncoder
# Filter only Final matches
Finals_df=df[df['Match_Type']=='Final']

# Encode Toss Winner and Match Winner as numbers
le=LabelEncoder()
Finals_df['Toss_Winner_Code']=le.fit_transform(Finals_df['Toss_Winner'])
Finals_df['Winner_Code']=le.fit_transform(Finals_df['Winner'])

# Scatter Plot
plt.figure(figsize=(10, 6))
plt.scatter(Finals_df['Toss_Winner_Code'],Finals_df['Winner_Code'],color='crimson',s=100,alpha=0.7,edgecolors='k')
plt.title('Toss Winner vs Match Winner (IPL Finals)',fontsize=16)
plt.xlabel('Toss Winner (encoded)',fontsize=14)
plt.ylabel('Match Winner (encoded)',fontsize=14)
plt.grid(True,linestyle='--',alpha=0.7)

plt.xticks(ticks=range(len(le.classes_)),labels=le.classes_, rotation=90)
plt.yticks(ticks=range(len(le.classes_)),labels=le.classes_)

plt.tight_layout()
plt.show()     #To Plot the Graph
```

### Q.12) How can you create a stacked bar chart to show the frequency of Super Over occurrences by match type?

```python
# Create a table
Super_Over_Table=pd.crosstab(df['Match_Type'],df['Super_Over'])

# Plot the graph
Super_Over_Table.plot(kind='bar',stacked=True,figsize=(10,6),color=['skyblue','salmon'])
plt.title('Super Over Frequency by Match Type')
plt.xlabel('Match Type')
plt.ylabel('Number of Matches')
plt.show()
```

### Q.13) What is the Performance of Teams after Winning Toss?

```python
# Create a mask where Toss_Winner==Winner
df['Toss_Win_and_Match_Win']=(df['Toss_Winner']==df['Winner'])

# Group by toss_winner and calculate success rate
Team_Success=df.groupby('Toss_Winner')['Toss_Win_and_Match_Win'].mean().sort_values(ascending=False)

sns.heatmap(Team_Success.to_frame(),annot=True,cmap='YlGnBu',cbar=False)    # Plot the graph (HeatMap)
plt.title('Team Success % After Winning Toss')
plt.show()
```

### Q.14) What is the relation between Teams Vs Team Win % ?

```python
# Create a helper column
df['Team1_Win']=df['Winner']==df['Team_1']

# Create a pivot table: Team 1 wins vs Team 2
pivot_table = pd.pivot_table(
    df,
    index='Team_1',
    columns='Team2',
    values='Team1_Win',
    aggfunc='sum',
    fill_value=0
)

# Plot the graph (HeatMap)
plt.figure(figsize=(12,10))
sns.heatmap(pivot_table, annot=True, fmt='d', cmap='coolwarm')
plt.title('Team 1 vs Team 2 Wins')
plt.xlabel('Team 2')
plt.ylabel('Team 1')
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.show()
```
